```python
from ultralytics import YOLO

# Load the nano model
model = YOLO('yolo11n.pt')

# Train with specific settings for high accuracy
model.train(
    data=f"{dataset.location}/data.yaml", # Path from your Roboflow cell
    epochs=100,
    imgsz=640,
    batch=-1,              # -1 tells YOLO to auto-detect the best batch size
    optimizer='AdamW',     # Excellent for fine-tuning
    patience=20,           # Stop early if the model stops improving
    save=True,
    device=0,
    # Augmentation inside the code
    mosaic=1.0,
    mixup=0.1,
    box=7.5,               # Increased weight on box accuracy
    cls=0.5                # Weight on classification
)
```
